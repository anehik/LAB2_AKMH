```python
sesgos['resultados']
```

Definiciones

- status_quo:  % de ocurrencias donde capital_perdedora/capital_acm < capital_ganadora/capital_acm.

- aversion_perdida: % de ocurrencias donde capital_perdedora/capital_ganadora es > 1.5 

- sensibilidad_decreciente: Comparar lo siguiente para la primera operación "ganadora" y la última operación "ganadora":
    Si el capital_acm de la cuenta aumentó.
    Si el capital_ganadora y/o capital_perdedora aumentaron (cualquiera de los dos, pero, al menos uno).
    Si el ratio capital_perdedora/capital_ganadora sigue siendo > 1.5.
    Si y sólo si todas las anteriores se cumplen, entonces, SI hay sensibilidad decreciente.
